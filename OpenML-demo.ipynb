```python
import openml
import pandas as pd
```

```python
datasets_list = openml.datasets.list_datasets()

datalist = pd.DataFrame.from_dict(datasets_list, orient='index')
datalist = datalist[[
    'did', 'name', 'NumberOfInstances',
    'NumberOfFeatures', 'NumberOfClasses'
]]

datalist.head(10)
```

```python
datalist[datalist.NumberOfInstances > 10000].sort_values(['NumberOfInstances']).head(10)
```

```python
datalist[datalist.NumberOfFeatures > 80].sort_values(['NumberOfFeatures']).head(10)
```

```python
datalist.query('NumberOfFeatures > 20 and NumberOfClasses < 5').head(10)
```
